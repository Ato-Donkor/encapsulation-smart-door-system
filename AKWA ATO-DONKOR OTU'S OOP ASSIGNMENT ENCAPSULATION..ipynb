
## Encapsulation: Smart Door Access System


```python
class Staff:
    """
    Encapsulated staff access record for the UMaT Smart Door System.
    """

    def __init__(self, s_name, access_code):
        self.s_name = s_name
        self.__access_code = access_code
        self.__unlocked = False

    def unlock(self, passphrase):
        """
        Unlock the access code using the admin passphrase.
        """
        if passphrase == "umat_secure_2024":
            self.__unlocked = True
            print(f"🔓 {self.s_name}'s record has been unlocked.")
            return True
        print(f"🔒 Unlock failed for {self.s_name}. Incorrect passphrase.")
        return False

    @property
    def access_code(self):
        """
        Retrieve the access code only if the record is unlocked.
        """
        if not self.__unlocked:
            return "[LOCKED — Admin passphrase required to view]"
        return self.__access_code

    @access_code.setter
    def access_code(self, code):
        """
        Update the access code with minimum length validation.
        """
        if len(code) < 4:
            raise ValueError(
                f"The code '{code}' is too short. It must be at least 4 characters."
            )
        self.__access_code = code
        print(f"✅ Access code for {self.s_name} updated successfully.")

    def print_details(self):
        """
        Print the staff record with the access code concealed.
        """
        hidden = "■" * len(self.__access_code)
        print("┌" + "─" * 40 + "┐")
        print(f"│{'STAFF ACCESS RECORD':^40}│")
        print("├" + "─" * 40 + "┤")
        print(f"│  Staff Name  : {self.s_name:<25}│")
        print(f"│  Code        : {hidden:<25}│")
        print(f"│  Lock Status : {'UNLOCKED' if self.__unlocked else 'LOCKED':<25}│")
        print("└" + "─" * 40 + "┘")
```

## Running the Demonstration

```python
staff = Staff("Dr. Matthew Cobbinah", "UMAT54")
staff.print_details()

print("\n--- Attempt without unlocking ---")
print(staff.access_code)

print("\n--- Wrong passphrase ---")
staff.unlock("wrong_pass")

print("\n--- Correct passphrase ---")
staff.unlock("umat_secure_2024")
print(f"Access Code: {staff.access_code}")

print("\n--- Reject short code ---")
try:
    staff.access_code = "Hi"
except ValueError as e:
    print(f"Error: {e}")

print("\n--- Accept valid code ---")
staff.access_code = "UMAT2024"
staff.print_details()
```
